```python
import pandas as pd
import numpy as np
```

```python
rango1 = np.linspace(1,50)
rango2 = np.linspace(51,150,100)
```

```python
# ejecuta un loop que devuelva si el numero en el primer array es par o impar
for i in rango1:
    if i % 2 == 0:
        print(i, "es par")
    else:
        print(i, "es impar")
```

```python
counts = [0] * 4
for i in rango2:
    if i % 2 == 0 or i % 3 == 0:
        # print(i, " es divisible por 2 o 3")
        counts[0] += 1
    if i % 2 == 0 and i % 3 == 0:
        # print(i, " es divisible por 2 y 3")
        counts[1] += 1
    if i % 2 != 0 and i % 3 == 0:
        # print(i, " es divisible por 3 pero no por 2")
        counts[2] += 1
    if i % 2 != 0 and i % 3 != 0:
        # print(i, " no es divisible ni por 3 ni por 2")
        counts[3] += 1

print(f"""
    La cantidad de numeros divisibles por 2 o 3 es: {counts[0]}
    La cantidad de numeros divisibles por 2 y 3 es: {counts[1]}
    La cantidad de numeros divisibles por 3 pero no por 2 es: {counts[2]}
    La cantidad de numeros no divisibles ni por 3 ni por 2 es: {counts[3]}
    """)
```

```python
df['underperforming'] = 0
```

```python
df.underperforming = (df.passengers < df.passengers.mean()) * 1
```

```python
df[df.underperforming == 1]
```

```python
df.underperforming.value_counts()
```

```python
df_mean = df.passengers.mean()
df_std = df.passengers.std()
print(f'La media es {df_mean} y la desviación es {df_std}')
```

```python
df['outlier'] = 0
df.outlier = ((df.passengers > df_mean + df_std) * 1) + ((df.passengers < df_mean - df_std) * 1)
df.outlier.value_counts()
```
