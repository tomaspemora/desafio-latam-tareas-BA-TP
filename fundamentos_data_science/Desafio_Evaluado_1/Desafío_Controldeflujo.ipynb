### <mark>Requerimiento 1</mark>: Generar un par de arrays ficticios con numpy.   

```python
# Se importa la libreria numpy, y además se definen dos array utilizando numby (as np) y linspace para generar dos listas de acuerdo a la instrucción:
import numpy as np
import pandas as pd

array_1 = np.linspace(1,50)
array_2 = np.linspace(51,150,100)
print(array_1)
print(array_2)
```

### <mark>Requerimiento 2</mark>: Ejecuta un loop que devuelva si el número en el primer array es par o impar

```python
# se define un ciclor for y las condiciones necesarias para determinar que elementos son pares/impares de nuestra lista "array_1":
for i in array_1:
    if i % 2 == 0:
        print(f"{i} es par ")
    else:
        print(f"{i} es impar")
```

### <mark>Requerimiento 3</mark>: Genere un loop con el segundo array que cuente las siguientes condiciones:
    ● Si el número es divisible por 2 o 3.
    ● Si el número es divisible por 2 y 3.
    ● Si el número es divisible por 3 pero no por 2.
    ● si el número no es divisible por 2 ni 3.

```python
#se define un código que permita ejecutar cada una de las condiciones indicadas anteriormente para nuestra lista "array_2"
counts = [0] * 4
for i in array_2:
    if i % 2 == 0 or i % 3 == 0:
        # print(i, " es divisible por 2 o 3")
        counts[0] += 1
    if i % 2 == 0 and i % 3 == 0:
        # print(i, " es divisible por 2 y 3")
        counts[1] += 1
    if i % 2 != 0 and i % 3 == 0:
        # print(i, " es divisible por 3 pero no por 2")
        counts[2] += 1
    if i % 2 != 0 and i % 3 != 0:
        # print(i, " no es divisible ni por 3 ni por 2")
        counts[3] += 1

print(f"""
    La cantidad de numeros divisibles por 2 o 3 es: {counts[0]}
    La cantidad de numeros divisibles por 2 y 3 es: {counts[1]}
    La cantidad de numeros divisibles por 3 pero no por 2 es: {counts[2]}
    La cantidad de numeros no divisibles ni por 3 ni por 2 es: {counts[3]}
    """)
```

### <mark>Requerimiento 6</mark>: Por último, haz un loop que clasifique los meses donde la cantidad de pasajeros se escapa de la tendencia. Para ello:
    ● Genera dos objetos que guarden la media general y la desviación estándar general de passengers.
    ● Genera una nueva columna en la tabla de datos que se llame outlier y asígnale 0.
    ● Ejecuta un loop que recorra cada observación de passengers, donde si la observación se escapa de la tendencia sea 1, de lo contrario 0.
    ● Para clasificar los casos que se escapen la tendencia, la observación debe satisfacer una de las siguientes condiciones:
        ○ La observación debe ser menor a la media menos la desviación estándar, o
        ○ La observación debe ser mayor a la media más la desviación estándar.
        ○ ¿Cuántas observaciones se pueden clasificar como casos extremos?

```python
#Se parte por generar los dos objetos que guarden la media general y la desviación estándar de la columna passengers:
df_mean = df.passengers.mean()
df_std = df.passengers.std()
print(f'La media general es {df_mean} y la desviación estándar es {df_std}')
```

```python
#Se genera nueva columna llamada "outlier" y se le asigna valor 0:
df["outlier"] = 0
```

```python
df.head() #Corroborarmos la existencia de nuestra nueva columna "ouliter" con valores 0:
```

```python
df.outlier = ((df.passengers > df_mean + df_std) * 1) + ((df.passengers < df_mean - df_std) * 1) #En columna outlier se asigna valores 1 de acuerdo a las condiciones indicadas:
df.outlier.value_counts() #Se corrobora la cantidad de valores 0 y 1 de la columna outlier:
```
